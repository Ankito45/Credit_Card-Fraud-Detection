```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# loading the dataset to a Pandas DataFrame
credit_card_data = pd.read_csv('creditcard.csv')
```

```python
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
# gives the complete dataset information
credit_card_data.info()
```

```python
# checking whether the data contain any null/missing values or not in each column
credit_card_data.isnull()
```

```python
# distribution of legit and fraud transaction
# here 0 -> legit transaction
#      1 -> fraudlent transaction
credit_card_data['Class'].value_counts()
```

```python
# distributing the values for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# statistical measure of the amount column fro mkre analysis
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values fro both transaction
credit_card_data.groupby('Class').mean()
```

```python
# since the data in the legit is very big so taking a random sample of size  
legit_sample = legit.sample(n=492)
```

```python
# now conacatenate two dataset 
new_datasheet = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_datasheet.head()
```

```python
new_datasheet.tail()
```

```python
new_datasheet['Class'].value_counts()
```

```python
# compare the values fro both transaction
new_datasheet.groupby('Class').mean()
```

```python
x = new_datasheet.drop(columns='Class', axis = 1)
y = new_datasheet['Class']
```

```python
print(x)
```

```python
print(y)
```

```python
# spliting the data for training and testing data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, stratify = y, random_state = 2)
```

```python
# checking whether the data splitted or not
print(x.shape, x_train.shape, x_test.shape)
```

```python
# For testing the model training is used as LogisticRegression
model = LogisticRegression()
```

```python
# training the Logistic Regression Model with Training Data
model.fit(x_train, y_train)
```

```python
# accuracy on training data is -> Model Evaluation 
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction, y_train)
```

```python
print('Accuracy on Training data : ',training_data_accuracy)
```

```python
# accuracy on test data
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction, y_test)
```

```python
print('Accuracy on Test data :',test_data_accuracy)
```
